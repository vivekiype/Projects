```python
def graph(y):
    sns.boxplot(x="Classification", y=y, data=data)
    plt.figure(figsize=(10,10))

plt.subplot(221)
graph('SL')
 
plt.subplot(222)
graph('SW')
 
plt.subplot(223)
graph('PL')
 
plt.subplot(224)
graph('PW')
 
plt.show()
```
